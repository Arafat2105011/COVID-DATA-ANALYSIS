```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_40_Deaths = data.nlargest(40,'Deaths')

plt.figure(figsize=(15, 7))
sns.barplot(x='Region', y='Confirmed', data=top_40_Deaths)
plt.title('Confirmed Cases by Region')
plt.xlabel('Region')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=40)
plt.show()
```

```python
top_25_country=data.nlargest(25, 'Mortality Rate')

data['Mortality Rate'] = (data['Deaths'] / data['Confirmed']) * 100
plt.figure(figsize=(13, 6))
sns.barplot(x='Region', y='Mortality Rate', data=top_25_country)
plt.title('Mortality Rate by Region')
plt.xlabel('Region')
plt.ylabel('Mortality Rate (%)')
plt.xticks(rotation=55)
plt.show()
```

```python
corr_matrix = data[['Confirmed', 'Deaths', 'Recovered']].corr()
plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
